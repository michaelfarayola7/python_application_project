### Choosing an optimizer and loss

Keep in mind what kind of problem you are trying to solve:

    # For a multi-class classification problem
    model.compile(optimizer='rmsprop',
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])

    # For a binary classification problem
    model.compile(optimizer='rmsprop',
                  loss='binary_crossentropy',
                  metrics=['accuracy'])

    # For a mean squared error regression problem
    model.compile(optimizer='rmsprop',

## Scaling and Train Test Split

```python
X = df.drop('price',axis=1).values
y = df['price'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=101)
```

##### Scaling

NB: You only fit and transform the X_train while you only transform X_test to avoid data leakage

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
X_train.shape
```

#### NB: We set the initial dense layer based on the number of incoming features

```python
model = Sequential()

model.add(Dense(19, activation = 'relu'))
model.add(Dense(19, activation = 'relu'))
model.add(Dense(19, activation = 'relu'))
model.add(Dense(19, activation = 'relu'))

model.add(Dense(1))

model.compile(optimizer = 'adam', loss = 'mse')
```

```python
# It's very typical to do batch sizes in powers of two, so 64, 128 etc
```

```python
model.fit(x=X_train, y=y_train, validation_data=(X_test, y_test), batch_size = 128, epochs=400)
```

```python
losses = pd.DataFrame(model.history.history)
```

##### In a situation where we see the val_loss increasing, then it shows our model is overfitting

```python
losses.plot()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, explained_variance_score
```

```python
prediction = model.predict(X_test)
```

```python
prediction
```

```python
mean_squared_error(y_test, prediction)
```

```python
np.sqrt(mean_squared_error(y_test, prediction))
```

```python
mean_absolute_error(y_test, prediction)
```
